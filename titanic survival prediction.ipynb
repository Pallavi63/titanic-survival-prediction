```python
#droping embarked, Ticket, passengerID, and Name 
titanic_df.drop(['Name', 'Ticket', 'Embarked', 'PassengerId'], axis = 1, inplace = True)
```

```python
#the average age for male (~29) and female (~25)
plt.figure(figsize=(15, 10))
sns.boxplot(x = 'Sex', y = 'Age', data = titanic_df)
```

```python
#only only column to represent male and female 
pd.get_dummies(titanic_df['Sex'])
```

```python
male = pd.get_dummies(titanic_df['Sex'], drop_first = True)
```

```python
#drop the embarked and sex 
titanic_df.drop(['Sex'], axis = 1, inplace = True)
```

```python
titanic_df
```

```python
#add encoded column male again
titanic_df = pd.concat([titanic_df, male], axis = 1)
```

```python
titanic_df
```

```python
#drop the target coloumn before we do train test split
X = titanic_df.drop('Survived', axis = 1).values
y = titanic_df['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_predict_test = classifier.predict(X_test)
y_predict_test
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot = True, fmt = "d")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_test))
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
y_predict_test = classifier.predict(X_test)

cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```
